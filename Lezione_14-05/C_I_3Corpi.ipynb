# Condizioni Iniziali Problema 3 Corpi

## Eulero

Nel caso del problema dei tre corpi di Eulero un set di posizioni e velocità iniziali è il seguente:

```python
pos_Ax = -1
pos_Ay = -1

pos_Bx = 0
pos_By = 0

pos_Cx = 1
pos_Cy = 0

Vel_Ax = 0
Vel_Ay = -1

Vel_Bx = 0
Vel_By = 0

Vel_Cx = 0
Vel_Cy = 1
```

Ovviamente questi valori possono essere scalati di un fattore moltiplicativo che lascia invariate le proporzioni

## Lagrange

Nel caso del problema dei tre corpi di Lagrange un set di posizioni e velocità iniziali è il seguente:

```python
pos_Ax = 0
pos_Ay = 0

pos_Bx = 1
pos_By = 0

pos_Cx = 1/2
pos_Cy = 3**0.5 / 2

Vel_Ax = 1/2
Vel_Ay = -3**0.5 / 2

Vel_Bx = 1/2
Vel_By = 3**0.5 / 2

Vel_Cx = -1
Vel_Cy = 0
```

Ovviamente questi valori possono essere scalati di un fattore moltiplicativo che lascia invariate le proporzioni
